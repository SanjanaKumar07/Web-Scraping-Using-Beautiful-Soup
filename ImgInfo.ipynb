```python
from selenium import webdriver
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import urllib.request
import time
import sys
import os

#Information

html=urlopen("https://www.commandonetworks.com/switches/SF50-5")
#taking user input
bsObj=BeautifulSoup(html.read())
name=bsObj.find("h2",{"class":"post-title"})
print(name.get_text())
description =bsObj.find("div",{"class":"block-devider"})
print(description.get_text())
info =bsObj.find("div",{"class":"page-section hero"},{"id":"info"})
print(info.get_text())
specification =bsObj.findAll("tr")
for tag in specification:
    print(tag.get_text())
    
    
#Image Download

site = "https://www.commandonetworks.com/switches/SF50-5"
#providing driver path
driver = webdriver.Firefox(executable_path = 'C:\Drivers\geckodriver.exe')
#passing site url
driver.get(site)
#if you just want to download 10-15 images then skip the while loop and just write
#driver.execute_script("window.scrollBy(0,document.body.scrollHeight)")
#below while loop scrolls the webpage 7 times(if available)
#parsing
soup = BeautifulSoup(driver.page_source, 'html.parser')
#closing web browser
driver.close()
#scraping image urls with the help of image tag and class used for images
img_tags = soup.find_all("img", class_="product-v-img" )
count = 0
for i in img_tags:
    #print(i['src'])
    try:
#passing image urls one by one and downloading
        urllib.request.urlretrieve(i['src'], str(count)+".jpg")
        count+=1
        print("Number of images downloaded = "+str(count),end='\r')
    except Exception as e:
        pass
```
