```python
from urllib.request import urlopen
```

```python
from bs4 import BeautifulSoup
```

```python
html=urlopen("https://www.commandonetworks.com/switches/SF50-5")
```

```python
bsObj=BeautifulSoup(html.read())
print(bsObj.prettify())
```

```python
name=bsObj.find("h2",{"class":"post-title"})
print(name.get_text())
description =bsObj.find("div",{"class":"block-devider"})
print(description.get_text())
info =bsObj.find("div",{"class":"page-section hero"},{"id":"info"})
print(info.get_text())
specification =bsObj.findAll("tr")
for tag in specification:
    print(tag.get_text())

```
